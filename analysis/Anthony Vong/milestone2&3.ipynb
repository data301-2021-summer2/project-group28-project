```python
test = obesity[["Country","Obesity","Total_Active_Cases"]]
test
```

```python
test.describe(exclude=np.object).T
```

```python
import pandas_profiling

profile = pandas_profiling.ProfileReport(test)
display(profile)
```

#### The report has shown that the active covid-19 cases are highly correlated to the rate of obesity.

```python
test = test.assign(size = lambda x: np.where((x.Total_Active_Cases > 50000), "above","below"))
test
```

```python
order_test = (test
                 .groupby('size')["Obesity"]
                 .mean()
                 .round(2)
                 .sort_values(ascending=False)
                 .reset_index()
             )

order_test
```

#### We have found that the average rate of obesity for countries that have more than 10000 active cases is: 24.50%

#### On the other hand, the average rate of obesity for countries that have more than 10000 active cases is: 17.03%

```python
plt.subplots(figsize=(6, 4))

sns.boxplot(data=test,
           x='Obesity',
           y='size',
           order = order_test['size'],
           palette="vlag",
           width=0.8)

sns.stripplot(x="Obesity", y="size", data=test,size=4, color=".4", linewidth=0)

plt.ylabel('Active covid cases in countries \n (above / below 50000 cases)')
plt.xlabel('The rate of Obesity in different country \n during covid-19 pandemic')
plt.title('Comparing the prevalence of obesity with COVID-19 active cases.')
```

```python
sns.catplot(data=test, kind="bar", x="Obesity", y="size")

plt.ylabel('Active covid cases in countries \n (above / below 50000 cases)')
plt.xlabel('The rate of Obesity in different country \n during covid-19 pandemic')
plt.title('Comparing the prevalence of obesity with COVID-19 active cases.')
```

```python
sns.displot(data=test, kind="ecdf", x="Obesity", rug=True)

plt.ylabel('Active covid cases in countries \n (measurment: one million)')
plt.xlabel('The rate of Obesity in different country \n during covid-19 pandemic')
plt.title('Comparing the prevalence of obesity with COVID-19 active cases.')
```

```python
sns.set_theme(style="white")

sns.relplot(x="size", y="Obesity",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=6, data=test)

plt.ylabel('The rate of Obesity in different country \n during covid-19 pandemic')
plt.xlabel('Active covid cases in countries \n (above / below 50000 cases)')
plt.title('Comparing the prevalence of obesity with COVID-19 active cases.')
```

```python
sns.violinplot(data=test, x="Obesity", y="size",
               split=True, linewidth=1)
               
sns.despine(left=True)

plt.ylabel('Active covid cases in countries \n (above / below 50000 cases)')
plt.xlabel('The rate of Obesity in two categories of countries ')
plt.title('Comparing the prevalence of obesity with COVID-19 active cases.')
```

## Research topic: Comparing the prevalence of obesity with COVID-19 active cases.


#### According to the results, we are confident to say that active covid-19 cases have very high correlation with the rate of obesity. The countries that have more than 50000 active covid-19 cases tend to have higher rate in obesity.


### The reasons could be: 
#### Those countries (has more than 50000 active cases) are excuting the lockdown / strict social distancing policy, which results in people have much less time / no chance to excerise.
