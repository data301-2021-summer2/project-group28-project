```python
fat1 = fat[['Country','Obesity','Confirmed','Deaths','Recovered','Active','Population']]
fat1
```

```python
kcal['Confirmed cases in count'] = kcal['Population']*kcal['Confirmed']
kcal['Confirmed cases in count']
```

```python
kcal['Active cases in count'] = kcal['Population']*kcal['Active']
kcal['Active cases in count']
```

```python
kcal1 = pd.concat([kcal['Confirmed cases in count'], kcal['Active cases in count']], axis = 1)
kcal1
```

```python
table = pd.concat([fat1, kcal1], axis = 1)
table
```

```python
table_set1 = table.iloc[:,[7,8]].div(100).round()
table_set1
```

```python
obesity_1 = pd.concat([fat1, table_set1], axis = 1)
obesity_1
```

```python
obesity = obesity_1.dropna().reset_index(drop=True)
obesity
```
