## Method Chaining

```python
def load_and_process(fat_path, protein_path) :
    
    df1 = pd.read_csv(fat_path)[['Country','Animal Products','Deaths']].rename(columns = {
    'Animal Products' : '%Fat in Animal Products',
    'Deaths':'% COVID-19 Deaths'
    }).dropna().reset_index().drop(['index', 'Country','% COVID-19 Deaths'],axis =1)
    
    
    df2 = pd.read_csv(protein_path)[['Country','Animal Products','Deaths']].rename(columns = {
    'Animal Products' : '% Protein in Animal Products',
    'Deaths':'% COVID-19 Deaths'
    }).dropna().reset_index().drop(['index'], axis =1)
    
    df3 = pd.concat([df2,df1],axis =1).reindex(columns = ['Country','% Protein in Animal Products', '%Fat in Animal Products', '% COVID-19 Deaths'])
    return df3

load_and_process("/Users/izgikocacelebi/Downloads/Fat_Supply_Quantity_Data.csv",'/Users/izgikocacelebi/Downloads/Protein_Supply_Quantity_Data.csv')
```
